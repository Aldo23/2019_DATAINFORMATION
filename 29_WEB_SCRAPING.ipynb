```python
from bs4 import BeautifulSoup
import requests 

import pandas as pd
from pandas import Series, DataFrame
```

```python
url='https://www.ucop.edu/operating-budget/budgets-and-reports/legislative-reports/2013-14-legislative-session.html'
```

```python
#we will convert this table in a pandas dataframe and than send a request to grab content 
#and convert in a pandas dataframe with bs
```

```python
result = requests.get(url)
c = result.content #now we have a bs object
```

```python
soup = BeautifulSoup(c)
```

```python
#if you check the caractheristics of this page the table is inside the html in a location that is called
#id="content"
```

```python
#create a dictionary of the class and id /// go to the section of interest
summary = soup.find('div',{'class':'list-land','id':'content'}) 
#find tables in html
tables = summary.find_all('table') #class="table-striped" ---see bs webpage for find_all operations 
tables
```

```python
data = []
rows = tables[0].findAll('tr') #tr table rows

rows
```

```python
for tr in rows:
    cols = tr.findAll('td') # table cells td - will find every cell 
    #print(cols)
    
    for td in cols: #check if text is in the row now
        text = td.find(text=True)
        print(text)
        data.append(text)
    
```

```python
data
```

```python
#now we set the pandas dataframe and only cells with the pdf file and the day of the report
```

```python
reports = []
date = []

index = 0

for item in data:
    if 'pdf' in item: # I know it's a report 
        date.append(data[index-1]) #append data which is always one position before the pdf
        
        reports.append(item.replace(u'\xa0',u' ')) #fix unicode 
        
    index +=1
```

```python
reports
```

```python
date
```

```python
date = Series(date)
reports = Series(reports)
```

```python
legislative_df = pd.concat([date,reports],axis=1)
```

```python
legislative_df.columns = ['Date','Report']
```
