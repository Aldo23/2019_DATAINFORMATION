```python
#seaborn comes with some datasets
tips = sns.load_dataset('tips')
tips
```

```python
tips.head()
```

```python
#tip left on a bill
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
#multplie regression plot with scatterd plot and linear plot separetly customized
# order 1 is linear
sns.lmplot('total_bill','tip',tips,order=1,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
#if you don't want linear fit you can specify the order you want to use
sns.lmplot('total_bill','tip',tips,order=4,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
#in this case we get rid of the linear regression
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
tips.head()
```

```python
#calculate percentage of the tip based on the total bill
tips['tip_pect']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_pect',tips)
```

```python
#we can add a jitter to make it easier to read
url='https://en.wikipedia.org/wiki/Jitter'
```

```python
sns.lmplot('size','tip_pect',tips,x_jitter=.1)
```

```python
#estimate the tendency of each bin -the size of each party-
```

```python
#this show the estimate depending of the tip percentage for the size of the party 
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
#who tips more female or male?
sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['x','o'])
```

```python
#check if the day makes difference and how much the tip is 
sns.lmplot('total_bill','tip_pect',tips,hue='day')
```

```python
#local regression
sns.lmplot('total_bill','tip_pect',tips,lowess=True,line_kws={'color':'black'})
```

```python
sns.regplot('total_bill','tip_pect',tips)#more robust plotting function always in regression plots
```
