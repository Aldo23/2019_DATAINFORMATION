```python
dataset1 = randn(100)
```

```python
plt.hist(dataset1)
```

```python
#10 bins is the default number of bins equally spaced
```

```python
dataset2 = randn(80)
plt.hist(dataset2, color='indianred')
```

```python
#if you want to see both histogram on the same plot
#we normalize the code because both graphs have different amount of numbers
#While the shape of a histogram tells us quite a bit, frequency as a value on the y-axis. 
#Changing the y axis values without changing the shape of the histogram is known as normalizing 
plt.hist(dataset1,density=True,color='indianred',alpha=0.5,bins=20)
plt.hist(dataset2,density=True,alpha=0.5,bins=20)
```
